```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x,y = np.meshgrid(np.linspace(-5,5,30),np.linspace(-5,5,30))

a = (1/((x-2)**2 + (y-2)**2))/(np.sqrt((x-2)**2 + (y-2)**2))
b = (1/((x+2)**2 + (y-2)**2))/(np.sqrt((x+2)**2 + (y-2)**2))
c = (1/((x-2)**2 + (y+2)**2))/(np.sqrt((x-2)**2 + (y+2)**2))
d = (1/((x+2)**2 + (y+2)**2))/(np.sqrt((x+2)**2 + (y+2)**2))
u = a*(x-2) + b*(x+2) - c*(x-2) - d*(x+2)
v = a*(y-2) + b*(y-2) - c*(y+2) - d*(y+2)

plt.quiver(x,y,u,v)
plt.show()
```
