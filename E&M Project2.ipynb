```python
#Non-custom imports.
#
#Imports for unused modules are commented out to minimize errors and setup.

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import math
#from matplotlib.animation import FuncAnimation
import time
#import matplotlib.animation as animation
#import seaborn as sns
#import pandas as pd
#from graphics import *
#import sys
import random
import re
from tqdm import tqdm
import pandas as pd
```

```python
n = 1
m = 1
a = 15
x = np.arange(0, a, 0.001)
y = np.sin(n*np.pi*x/a)
z = np.sin(m*np.pi*x/a)
yz = y*z

fig, axe = plt.subplots()
axe.plot(x, y, "RED")
plt.title(r"$\sin\left(\frac{n\pi x}{a}\right)$")
fig2, axe2 = plt.subplots()
axe2.plot(x, z, "BLUE")
plt.title(r"$\sin\left(\frac{m\pi x}{a}\right)$")
fig3, axe3 = plt.subplots()
axe3.plot(x, yz, "GREEN")
plt.title(r"$\sin\left(\frac{n\pi x}{a}\right)\sin\left(\frac{m\pi x}{a}\right)$")

num = 0
for i in range(len(x)):
    if (i != 0):
        dx = x[i]-x[i-1]
        y1 = yz[i-1]
        y2 = yz[i]
        num += dx*(y1 + (1/2)*(y2-y1)) #Trapezoid method

print("Sum: " + str(num))
print("Sum in terms of a: " + str(num/a) + "a")
```

```python
n = range(1, 8)
m = range(1, 8)
a = 15
x = np.arange(0, a, 0.00001)

sums = np.zeros([len(n), len(m)])

for i in range(len(n)):
    for j in tqdm(range(len(m))):
        y = np.sin(n[i]*np.pi*x/a)
        z = np.sin(m[j]*np.pi*x/a)
        yz = y*z
        
        num = 0
        for k in range(len(x)):
            if (k != 0):
                dx = x[k]-x[k-1]
                y1 = yz[k-1]
                y2 = yz[k]
                num += dx*(y1 + (1/2)*(y2-y1)) #Trapezoid method
        
        sums[i][j] = num/a

print(np.around(sums, 1))
```
