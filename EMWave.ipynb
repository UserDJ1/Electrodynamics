```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import math
from matplotlib.animation import FuncAnimation
import time
import matplotlib.animation as animation
import seaborn as sns
import pandas as pd
from graphics import *
import sys
import random
import math
```

```python
def lightwave(t0, tf, Δt, d, q, k=8.99e9, c=2.998e8):

    dt = (tf-t0)*Δt
    
    t = np.arange(t0, tf+dt, dt)
    
    d = 100+t*1000000000
    
    E = []
    for i in range(t.size):
        E.append(np.sin(t[i]))
    
    tdelayed = []
    for i in range(t.size):
        tdelayed.append(t[i]+((d[i])/c))

    #for i in range(len(tdelayed)):
    #    E.append(k*q/(d[i]-x[i])**2)
        
    return E, t, tdelayed, d
```

```python
E, t, tdelayed, d = lightwave(0, 10*math.pi, 0.001, 3000000000, 1)

fig, axe = plt.subplots()
axe.plot(t, E, "RED")
axe.plot(tdelayed, E, "BLUE")
```
